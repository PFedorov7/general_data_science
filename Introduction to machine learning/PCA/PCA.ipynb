```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('close_prices.csv')
data_test = pd.read_csv('djia_index.csv')
data.head()
```

```python
data_test.head()
```

```python
pca = PCA(n_components=10)
pca.fit(data.iloc[:,1:])
```

```python
sum(list(pca.explained_variance_ratio_)[0:4]) #4 components
pca.explained_variance_ratio_
```

```python
components = pca.transform(data.iloc[:,1:])
main_comp = []
for component in components:
    main_comp.append(component[0])
```

```python
DJI_list = data_test['^DJI'].tolist()
np.corrcoef(main_comp, DJI_list)
```
